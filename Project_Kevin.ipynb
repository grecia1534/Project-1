```python
import pandas as pd
import matplotlib.pyplot as plt

ks_zipcodes_df = pd.read_csv(r"resources/ks_zipcodes.csv")
mo_zipcodes_df = pd.read_csv(r"resources/mo_zipcodes.csv")
census_data_df = pd.read_csv(r"resources/census_data.csv")
ppp_data_df = pd.read_csv(r"resources/PPP_Data_Combined.csv")
```

```python
ppp_data_clean_df = ppp_data_df[["LoanRange","State","Zip","JobsReported","DateApproved"]]
ppp_data_clean_df = ppp_data_clean_df.rename(columns={"Zip":"Zipcode"})
ppp_data_clean_df.head()
```

```python
census_data_clean_df = census_data_df[["Zipcode","Population",
                                       "Poverty Count","Hispanic or Latino","White Alone"]]
census_data_clean_df.head()
```

```python
combo_clean_df = ppp_data_clean_df.merge(census_data_clean_df,on="Zipcode")
combo_clean_df["JobsReported"]=combo_clean_df["JobsReported"].fillna(0)
combo_clean_df = combo_clean_df.rename(columns={"LoanRange":"Range","JobsReported":"Jobs","DateApproved":"Date"})
combo_clean_df
```

```python
combo_clean_df.to_csv("combo_clean.csv")
```
